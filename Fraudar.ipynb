```python
import sys
import spartan as st
```

Using backend cpu

```python
# load graph data
tensor = st.loadTensor(path = "inputData/plain_graph_small.zip", col_types = [int, int], sep=' ')

stensor = tensor.toSTensor(hasvalue=False)
import pandas as pd
df = pd.DataFrame(tensor.data, columns=["user_id", "item_id"])

# 保存成 CSV（无索引、逗号分隔，标准格式）
df.to_csv("plain_graph_small.csv", index=False, encoding="utf-8")
```

```python
# create a anomaly detection model
fd = st.Fraudar(stensor)
```

```python
# run the model
fd.run(k=2, maxsize=[100, 100])
```
